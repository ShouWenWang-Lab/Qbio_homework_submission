```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("fluorescence_data.csv", delimiter=",", skiprows=1)

time_min = data[:, 0]          # 第一列：时间 (min)
fluorescence = data[:, 1]  # 第二列：荧光强度 (a.u.)
```

```python
plt.figure(figsize=(8,6))
plt.plot(time_min, fluorescence, marker='o', linestyle='-', color='blue', label="Fluorescence")

plt.xlabel("Time (minutes)", fontsize=12)
plt.ylabel("Fluorescence (a.u.)", fontsize=12)
plt.title("Gene Expression: Fluorescence vs. Time", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
max_fluorescence = np.max(fluorescence)

time_at_max = time_min[np.argmax(fluorescence)]

mask = (time_min >= 10) & (time_min <= 60)
time_window = time_min[mask]
fluorescence_window = fluorescence[mask]

rate = (fluorescence_window[-1] - fluorescence_window[0]) / (time_window[-1] - time_window[0])

std_dev = np.std(fluorescence_window)

print(f"Maximum fluorescence level: {max_fluorescence:.2f} a.u.")
print(f"Time at maximum fluorescence: {time_at_max:.2f} minutes")
print(f"Average rate of increase (10-60 min): {rate:.4f} a.u./min")
print(f"Standard deviation (10-60 min): {std_dev:.4f} a.u.")
```

## Interpretation of Results

The plot shows the fluorescence intensity over time, which reflects the dynamics of gene expression.  
- The **maximum fluorescence level** indicates the highest gene expression signal detected during the experiment.  
- The **time of maximum fluorescence** reveals when the expression reaches its peak.  
- The **average rate of fluorescence increase between 10–60 minutes** measures how quickly gene expression rises in the early/mid phase.  
- The **standard deviation in this interval** reflects the variability of gene expression levels during the active expression phase.  

Together, these metrics provide insight into both the dynamics and stability of the gene expression process.
