### Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Process

```python
data = np.loadtxt("/storage/daixiaoyunLab/gongyuntao/course/Qbio_homework_submission/module-1-problem/fluorescence_data.csv", delimiter=",", skiprows=1)
```

#### 1.  Fluorescence vs. Time

```python
time = data[:, 0]
fluorescence = data[:, 1]

plt.figure(figsize=(6,4), dpi=300)
plt.plot(time, fluorescence, marker='o', linestyle='-', color='blue',
         linewidth=2, markersize=6, label="Fluorescence")

plt.xlabel("Time (min)", fontsize=14)
plt.ylabel("Fluorescence (a.u.)", fontsize=14)
plt.title("Fluorescence vs. Time", fontsize=16, fontweight='bold')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

plt.show()
```

#### 2. Summary

```python
# 1. The maximum fluorescence level reached
max_fluorescence = np.max(fluorescence)

# 2. The time at which this maximum is reached
time_at_max = time[np.argmax(fluorescence)]

# 3.The average rate of fluorescence increase between t=10 and t=60 minutes
f10 = fluorescence[time == 10][0]
f60 = fluorescence[time == 60][0]
avg_rate = (f60 - f10) / (60 - 10)

# 4. The standard deviation of the data between t=10 and t=60 mins
mask = (time >= 10) & (time <= 60)
std_dev = np.std(fluorescence[mask])


# Out
print(f"Maximum fluorescence level: {max_fluorescence:.2f} a.u.")
print(f"Time at maximum fluorescence: {time_at_max:.2f} min")
print(f"Average rate of fluorescence increase (10–60 min): {avg_rate:.4f} a.u./min")
print(f"Standard deviation of fluorescence (10–60 min): {std_dev:.4f} a.u.")
```

#### 3. Figure interpretation


The plot shows the change in fluorescence intensity as a function of time.  
Since fluorescence here is used as a proxy for gene expression (e.g., reporter gene activity), the curve reflects how expression levels evolve over the experimental period.

- **Maximum fluorescence level**: This value indicates the peak expression of the gene during the experiment. It represents the highest activity reached by the promoter or reporter system.  
- **Time at maximum fluorescence**: This time point provides insight into when the expression system reaches its peak, which can be important for understanding promoter kinetics or cellular response times.  
- **Average rate of fluorescence increase (10–60 min)**: This metric quantifies how quickly gene expression ramps up during the early-to-mid phase of the experiment. A higher rate suggests stronger or faster induction of the gene.  
- **Standard deviation of fluorescence (10–60 min)**: This measures the variability in gene expression during the growth phase. A lower standard deviation indicates consistent expression, while a higher value suggests fluctuations in expression dynamics.

Overall, the data illustrate the temporal dynamics of gene expression, showing both the induction phase and eventual saturation.
